## Usage
### The YOLO command line interface (CLI) lets you simply train, validate or infer models on various tasks and versions. CLI requires no customization or code.

Example syntax:

```python
yolo task=detect    mode=train    model=yolov8n.yaml      args...
          classify       predict        yolov8n-cls.yaml  args...
          segment        val            yolov8n-seg.yaml  args...
                         export         yolov8n.pt        format=onnx  args...
```

```python
!yolo task=detect  mode=predict model=yolov8n.pt  source="test.jpg" save=True
```

```python
!yolo predict model=yolov8n.pt source='https://ultralytics.com/images/zidane.jpg'
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(data='config.yaml',epochs=15, device = 0)
```

```python
!yolo task=detect mode=val model='runs/detect/train/weights/best.pt' data=config.yaml
```

```python
!yolo task=detect mode=predict model='runs/detect/train/weights/best.pt' source="a.png" save=True
```
